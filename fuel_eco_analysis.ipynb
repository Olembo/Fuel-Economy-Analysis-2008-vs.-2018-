# 3. Cleaning Data

```python
# Removing unwanted columns
data_2008 = data_2008.drop(columns = ['Stnd', 'FE Calc Appr', 'Unadj Cmb MPG', 'Sales Area' ])
data_2018 = data_2018.drop(columns = ['Stnd', 'Stnd Description', 'Comb CO2', 'Cert Region'])
```

```python
#  columns names to lowercase and replace spaces with underscores
data_2008.rename(columns=lambda col: col.lower().replace(' ', '_'), inplace=True)
data_2018.rename(columns=lambda col: col.lower().replace(' ', '_'), inplace=True)
```

```python
# Remove duplicates
data_2008 = data_2008.drop_duplicates()
data_2018 = data_2018.drop_duplicates()
```

```python
# add Year column
data_2008['year'] = 2008
data_2018['year'] = 2018
```

```python
# unique values in the column
data_2008['cyl'].unique()
```

```python
# fixing Cyl column values
data_2008['cyl'] = data_2008['cyl'].replace({'(6 cyl)': 6,
                               '(4 cyl)': 4,
                               '(12 cyl)': 12,
                               '(8 cyl)': 8,
                               '(10 cyl)': 10,
                               '(16 cyl)': 16,
                               '(5 cyl)': 5,
                               '(2 cyl)': 2})
```

#### cleaning model column

```python
# Convert to Uppercase/Lowercase
def standardize_case(model_name):
    return model_name.upper()  # Or .lower()

data_2008['model'] = data_2008['model'].apply(standardize_case)
data_2018['model'] = data_2018['model'].apply(standardize_case)
```

```python
# Remove Extra Spaces
def remove_extra_spaces(model_name):
    return ' '.join(model_name.split())

data_2008['model'] = data_2008['model'].apply(remove_extra_spaces)
data_2018['model'] = data_2018['model'].apply(remove_extra_spaces)
```

```python
# Remove Special Characters
import re

def remove_special_chars(model_name):
    return re.sub(r'[^A-Za-z0-9\s]', '', model_name)  

data_2008['model'] = data_2008['model'].apply(remove_special_chars)
data_2018['model'] = data_2018['model'].apply(remove_special_chars)
```

### 6.3) Q: What are the characteristics of SmartWay vehicles? Have they changed over time? (mpg, greenhouse gas)

```python
# 1. Filter for SmartWay Vehicles
smartway_2008 = data_2008[data_2008['smartway'] == 'Yes']
smartway_2018 = data_2018[data_2018['smartway'] == 'Yes']
```

```python
# 2. Calculate Average MPG and Greenhouse Gas Score
avg_mpg_2008 = smartway_2008['cmb_mpg'].mean()
avg_mpg_2018 = smartway_2018['cmb_mpg'].mean()
avg_ghg_2008 = smartway_2008['greenhouse_gas_score'].mean()
avg_ghg_2018 = smartway_2018['greenhouse_gas_score'].mean()
```

```python
# 3. Compare and Analyse
print("SmartWay vehicle Characteristics:")
print(f"  Avg MPG 2008: {avg_mpg_2008: .2f}")
print(f"  Avg MPG 2018: {avg_mpg_2018: .2f}")
print(f"  Avg GHG Score 2008: {avg_ghg_2008: .2f}")
print(f"  Avg GHG Score 2018: {avg_ghg_2018: .2f}")

if avg_mpg_2018 > avg_mpg_2008:
    print("SmartWay vehicles have become more fuel-efficient.")
elif avg_mpg_2018 < avg_mpg_2008:
    print("SmartWay vehicles have become less fuel-efficient.")
else:
    print("SmartWay vehicles have not changed in fuel efficiency.")

if avg_ghg_2018 > avg_ghg_2008:
    print("SmartWay vehicles have become more environmentally friendly.")
elif avg_ghg_2018 < avg_ghg_2008:
    print("SmartWay vehicles have become less environmentally friendly.")
else:
    print("SmartWay vehicles have not changed in environmental impact.")
```

### 6.4) Q: What features are associated with better fuel economy (mpg)?

```python
# 1 Correlation Analysis
numerical_cols = ['displ', 'cyl', 'greenhouse_gas_score', 'air_pollution_score']
correlation_with_mpg = data_2018[numerical_cols + ['cmb_mpg']].corr()['cmb_mpg'].sort_values(ascending=False)
print("Correlation with MPG:\n", correlation_with_mpg)

# 2. Categorical Feature Analysis
categorical_cols = ['drive', 'smartway', 'trans', 'fuel', 'veh_class']
print("\nAverage MPG by Categorical Features:")
for col in categorical_cols:
    avg_mpg_by_category = data_2018.groupby(col)['cmb_mpg'].mean().sort_values(ascending=False)
    print(f"\n{col}:\n{avg_mpg_by_category}")
```

### 6.5) Q: For all of the models that were produced in 2008 that are still being produced in 2018, how much has the mpg improved and which vehicle improved the most?

```python
# 1. Identify common models (assuming 'model_name' column exists)
common_models = set(data_2008['model']).intersection(set(data_2018['model']))

# 2. Extract MPG data and calculate improvement
mpg_improvement = {}
for model in common_models:
    mpg_2008 = data_2008[data_2008['model'] == model]['cmb_mpg'].values[0]  # Get MPG for 2008
    mpg_2018 = data_2018[data_2018['model'] == model]['cmb_mpg'].values[0]  # Get MPG for 2018
    improvement = mpg_2018 - mpg_2008
    mpg_improvement[model] = improvement

# 3. Find the model with the maximum improvement
best_model = max(mpg_improvement, key=mpg_improvement.get)
max_improvement = mpg_improvement[best_model]

# 4. Print the results
print(f"The vehicle model with the largest MPG improvement is {best_model} with an improvement of {max_improvement:.2f} MPG.")
print("\nMPG Improvement for Common Models:")
for model, improvement in mpg_improvement.items():
    print(f"{model}: {improvement:.2f} MPG")
```
